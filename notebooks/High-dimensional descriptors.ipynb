```python
def sampleLandmarks(landmarks, number):
    configurations = [5, 9, 16, 21, 22, 27]
    if number not in configurations:
        raise Exception("Unknown configuration: %d landmarks"%number)
        
    if number == 5:
        landmarks_indexes = [[37, 38, 40, 41], [43, 44, 46, 47], 30, 48, 54]
    elif number == 9:
        landmarks_indexes = [36, 39, 45, 42, 30, 31, 35, 48, 54]
    elif number == 16:
        landmarks_indexes = [17, 21, 22, 26, 36, [37, 38, 40, 41], 39, 45, [43, 44, 46, 47], 42, 30, 31, 35, 48, 54, [62, 66]]
        #landmarks_indexes = [17, 21, 22, 26, 36, (37, 38, 40, 41), 39, 45, (43, 44, 46, 47), 42, 30, 31, 35]
    elif number == 21:
        landmarks_indexes = [17, 19, 21, 22, 24, 26, 27, 36, [37, 38, 40, 41], 39, 45, [43, 44, 46, 47], 42, 29, 31, 33, 35, 48, 51, 54, 57]
    
    if landmarks.shape[0] == 51:
        for i in range(len(landmarks_indexes)):
            if type(landmarks_indexes[i]) is list:
                for j in range(len(landmarks_indexes[i])):
                    landmarks_indexes[i][j] -= 17
            else:
                landmarks_indexes[i] -= 17
    
    extracted_landmarks = np.empty((number, 2))
    for i, indexes in enumerate(landmarks_indexes):
        if type(indexes) is list:
            extracted_landmarks[i] = np.mean(landmarks[indexes], axis=0)
        else:
            extracted_landmarks[i] = landmarks[indexes]
            
    return extracted_landmarks


def scaleImage(image, landmarks, scale):
    scales = [300, 212, 150, 106, 75]
    ratio = image.shape[1] / float(image.shape[0])
    return (cv2.resize(image, (int(ratio * scales[scale]), scales[scale])), landmarks * np.array((scales[scale], ratio*scales[scale])) / np.array(image.shape, dtype=np.float64))


def computeDescriptorAroundLandmark(descriptor, image, landmark, cells_size=(10,10), window_size=(4,4), normalize=True):
    height = window_size[0]*cells_size[0] + 2
    width = window_size[1]*cells_size[1] + 2
    x = int(landmark[0])-height/2
    y = int(landmark[1])-width/2

    if x >= image.shape[0] or x+height <= 0 or y >= image.shape[1] or y+width <= 0:
        cropped_image = np.zeros((height, width), dtype=np.uint8)
    elif x < 0 or x+height > image.shape[0] or y < 0 or y+width > image.shape[1]:
        cropped_image = np.zeros((height, width), dtype=np.uint8)
        cropped_image[-min(0,x):height-max(0, (x+height)-image.shape[0]), -min(0,y):width-max(0, (y+width)-image.shape[1])] = image[max(0,x):x+height, max(0,y):y+width]
    else:
        cropped_image = image[x:x+height, y:y+width]
    return descriptor.computeDescriptor(cropped_image, normalize=normalize)
```

```python
from itertools import izip, imap


def computeDescriptor(descriptor, image, landmarks, landmarks_number, scales, normalize):
    extracted_landmarks = sampleLandmarks(landmarks, landmarks_number)
    normalized_img, normalized_landmarks = normalizeImage(image, extracted_landmarks)
    desc = []
    
    for scale in scales:
        scaled_img, scaled_landmarks = scaleImage(normalized_img, normalized_landmarks, scale)
        for landmark in scaled_landmarks:
            desc.append(computeDescriptorAroundLandmark(descriptor, scaled_img, landmark.astype(np.int), normalize=normalize))
        
    return np.array(desc, dtype=np.float32).ravel()
        


def computeDescriptorsForAllImages(descriptor, images, landmarks, landmarks_number, scales, normalize):
    if landmarks_number is None:
        descs = np.empty((len(images), 7080), dtype=np.float32)
        for i, img in enumerate(images):
            descs[i] = descriptor.computeDescriptor(img[49:-49, 84:-84], normalize=False)
        return descs
    
    else:
        dim = landmarks_number*59*16*len(scales)
        descs = np.empty((len(images), dim), dtype=np.float32)
        for i, (img, lds) in enumerate(izip(images, landmarks)):
            descs[i] = computeDescriptor(descriptor, img, lds, landmarks_number, scales, normalize)
        return descs
```

```python
from scipy.sparse import lil_matrix
from sklearn.decomposition import PCA
from sklearn.random_projection import GaussianRandomProjection, SparseRandomProjection
from sklearn.lda import LDA


class AnalysisFramework:
    def __init__(self, ground_truth):
        self.descriptors = None
        self.compressed_descriptors = None
        self.ground_truth = ground_truth
        
    
    def computeStats(self, descriptors):
        scores = computeDistanceMatrix(descriptors, self.ground_truth)
        acc = computeAccuracy(scores)
        roc = computeROC(scores)
        return acc, roc
    
    
    def displayStats(self, labels, accs, rocs):
        for label, acc in zip(labels, accs):
            print "%s: %0.4f"%(label, acc)

        plotROC(rocs, labels, "roc")
    
    
    def computeDescriptors(self, descriptor_func, data):
        if self.descriptors is not None:
            del self.descriptors
            self.descriptors = None
            
        self.descriptors = descriptor_func(data)
        return self.computeStats(self.descriptors)
    
    
    def compressDescriptors(self, method="rp", dim=500, training_data=None):
        if self.compressed_descriptors is not None:
            del self.compressed_descriptors
            self.compressed_descriptors = None
        
        if method == "pca":
            self.compression = computeProbabilisticPCA(training_data, dim=dim)
            whitening = np.diag(np.power(self.compression.explained_variance_, -0.5))
        elif method == "rp":
            X = lil_matrix((len(self.descriptors), self.descriptors.shape[1]))
            self.compression = GaussianRandomProjection(n_components=dim)
            self.compression.fit(X)
            cov = np.cov(self.compression.transform(training_data), rowvar=0)
            _, s, v = np.linalg.svd(cov)
            #whitening = np.linalg.inv(np.dot(np.diag(np.sqrt(s)), v))
            whitening = np.eye(dim)
        elif method == "srp":
            X = lil_matrix((len(self.descriptors), self.descriptors.shape[1]))
            self.compression = SparseRandomProjection(n_components=dim, dense_output=True)
            self.compression.fit(X)
            cov = np.cov(self.compression.transform(training_data), rowvar=0)
            _, s, v = np.linalg.svd(cov)
            #whitening = np.linalg.inv(np.dot(np.diag(np.sqrt(s)), v))
            whitening = np.eye(dim)
        else:
            raise Exception("Compression method unknown")
        
        self.compressed_descriptors = self.compression.transform(self.descriptors)
        
        descriptors = np.copy(self.compressed_descriptors)
        for i in xrange(len(descriptors)):
            descriptors[i] = np.dot(whitening, descriptors[i])
        return self.computeStats(descriptors)
    
    
    def supervisedLearning(self, method="lda", dim=100, training_data=None):
        if method == "lda":
            supervised_learning = LDA(dim)
            supervised_learning.fit(training_data[0], training_data[1], True)
        else:
            raise Exception("Supervised learning method unknown")
            
        descriptors = supervised_learning.transform(self.compressed_descriptors)
        return self.computeStats(descriptors)
```

```python
class NullSpaceLDA:
    def __init__(self, dim=None):
        self.dim = dim
    
    
    def computeCovMatrices(self, x):
        k = len(x)
        classes_means = np.vstack((np.mean(x_k, axis=0) for x_k in x))
        xc = [x_k-mean_k for x_k, mean_k in zip(x, classes_means)]
        #s_w = np.sum([x_k.shape[0] * np.cov(x_k, rowvar=0) for x_k in x], axis=0)
        #s_b = k * np.cov(classes_means, rowvar=0)
        s_w = np.sum([np.dot(xc_k.T, xc_k) for xc_k in xc], axis=0)
        mean = np.mean(classes_means, axis=0)
        bincount = np.bincount(y)
        Xc = np.asarray([np.sqrt(count) * class_mean for count, class_mean in zip(bincount, classes_means)])
        s_b = np.dot(Xc.T, Xc)
        print s_w.shape, s_b.shape
        return s_w, s_b
    
    
    def fit(self, x, y):
        n = x.shape[1]
        classes = [x[y == k] for k in set(y)]
        s_w, s_b = self.computeCovMatrices(classes)
        r = np.linalg.matrix_rank(s_w)

        if r == n:
            print "Using normal LDA", s_w.shape, x.shape
            s_t = s_b + s_w
            m = np.dot(np.linalg.inv(mat(s_t)), s_b)
            _, s, v = np.linalg.svd(m)
            
        else:
            print "Using null space LDA"
            
            u, _, _ = np.linalg.svd(s_w)

            q = u[r:]
            trans = q.dot(np.transpose(q))
            print r
            s_b_tilde = trans.dot(s_b).dot(np.transpose(trans))

            _, s, v = np.linalg.svd(s_b_tilde)
        
        if self.dim:
            self.eigenvectors = np.transpose(v[:,:self.dim])
            self.eigenvalues = s[:self.dim]
        else:
            self.eigenvectors = np.transpose(v)
            self.eigenvalues = s
        
        self.mean = np.mean(x, axis=0)
        
        
    def transform(self, x):
        if len(x.shape) == 1:
            return self.eigenvectors.dot(x - self.mean)
        else:
            output = np.empty((x.shape[0], self.eigenvalues.shape[0]), dtype=np.float32)
            for i in range(x.shape[0]):
                output[i] = self.eigenvectors.dot(x[i] - self.mean)
            return output
        


def computeNullSpaceLDA(data, y, dim):
    lda = NullSpaceLDA(dim)
    lda.fit(data, y)
    return lda

        
def computeNullSpaceLDAs(data, y, sets, dim):
    ldas = []

    for i in range(len(sets)):
        training_set = []
        for k in range(len(sets)-1):
            training_set += sets[(i+k+1)%len(sets)]

        print "Computing LDA #%d"%(i+1)
        sys.stdout.flush()
        ldas.append(computeNullSpaceLDA(data[training_set], y[training_set], dim))

    return ldas


from sklearn.lda import LDA

def computeLDA(data, y, dim):
    print "using sklearn.lda"
    lda = LDA(dim)
    
    lda.fit(data, y, True)
    return lda
```
